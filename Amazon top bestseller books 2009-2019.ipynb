```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

# Import dataset

```python
data=pd.read_csv("bestsellers.csv")
data.head(10)
```

# Checking for missing values

```python
data.isnull().sum()
```

We can see that there is no missing value. Great!

```python
#Let's look for the correlation matrix to check which columns are correlated with each other
corr=data.corr()
```

```python
sns.heatmap(corr)
```

Below 0 shows that these columns are not affecting each other

# Top 10 books with highest number of ratings 2009-2019

```python
data[['Name','User Rating']].nlargest(10,'User Rating').plot.bar(x='Name',y='User Rating',color='purple')
plt.title('Top 10 books with highest number of ratings 2009-2019',size=15)
plt.xlabel('Name of the Books',fontsize=13,color='blue')
plt.ylabel('Ratings',size=13,color='blue')
plt.show()
```

# Top 10 books with highest reviews 2009-2019

```python
data[['Name','Reviews']].nlargest(10,'Reviews').plot.bar(x='Name',y='Reviews',color='coral')
plt.title('Top 10 books with highest reviews 2009-2019',fontsize=16)
plt.xlabel('Name of the Books',color='blue',fontsize=15)
plt.ylabel('Number of reviews',color='blue',fontsize=12)
plt.show()
```

# Most expensive book 2009-2019

```python
data[['Name','Price']].nlargest(10,'Price').plot.bar(x='Name',y='Price',color='green')
plt.title('Top 10 books with highest price 2009-2019',fontsize=16)
plt.xlabel('Name of the Books',color='blue',fontsize=15)
plt.ylabel('Price',color='blue',fontsize=15)
plt.show()   
```

# Top 10 Authors according to reviews

```python
data[['Author','Reviews']].nlargest(10,'Reviews').plot.bar(x='Author',y='Reviews',color='pink')
plt.title('Top 10 Authors according to reviews 2009-2019',size=15)
plt.xlabel('Authors',size=15,color='blue')
plt.ylabel('Number of reviews',size=15, color='blue')
plt.show()
```

# Top 10 Authors according to the price of the books 2009-2019

```python
data[['Author','Price']].nlargest(10,'Price').plot.bar(x='Author',y='Price',color='yellow')
plt.title('Top 10 Authors according to the price of the books 2009-2019',size=15)
plt.xlabel('Authors',size=15,color='blue')
plt.ylabel('Number of reviews',size=15, color='blue')
plt.show()
```
